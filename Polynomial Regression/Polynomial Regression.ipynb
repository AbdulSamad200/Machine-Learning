# Importing Necessary Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]

y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]
mymodel = np.poly1d(np.polyfit(x,y,3))
myline = np.linspace(2,95,100)
plt.scatter(x,y)
plt.plot(myline,mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y,mymodel(x))
```

```python
a = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
b =[100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
model = np.poly1d(np.polyfit(a,b,3))
line = np.linspace(1,22,100)
plt.scatter(a,b)
plt.plot(line,model(line))
plt.show()
```

```python
print(r2_score(b,model(a)))
```

```python
# Prediction
model(18)
```

# Polynomial Regression
#### Polynomial Regression equation:         y= b0+b1x + b2x2+ b3x3+....+ bnxn  
Polynomial Regression is a regression algorithm that models the relationship between a dependent(y) and independent variable(x) as nth degree polynomial.

### Steps for Polynomial Regression:

    The main steps involved in Polynomial Regression are given below:

      1.Data Pre-processing
      2.Build a Linear Regression model and fit it to the dataset
      3.Build a Polynomial Regression model and fit it to the dataset
      4.Visualize the result for Linear Regression and Polynomial Regression model. 
      5.Predicting the output.

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly= poly_reg.fit_transform(x)
pol_reg =  LinearRegression()
pol_reg.fit(X_poly,y)
```

```python
# Visualizing the Polynomial Regression Results
plt.scatter(x,y,color='red')
plt.plot(x,pol_reg.predict(poly_reg.fit_transform(x)),color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
    
```

```python
pred_linear =lin_Reg.predict([[10]])
```

```python
pred_polynomial= pol_reg.predict(poly_reg.fit_transform([[10]]))
```

```python
# Differennce between the prediction of Linnear Regression And Polynomial Regression Model
print("Linear Regression Model Result = ",pred_linear)
print("Polynomial regression Model result = ",pred_polynomial)
print("The difference is = ",pred_linear-pred_polynomial)
```
