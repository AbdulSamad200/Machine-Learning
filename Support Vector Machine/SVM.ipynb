```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
cancer =datasets.load_breast_cancer()
```

```python
#Print the name of 13 features
print("Features  = ",cancer.feature_names)

#Print the label type of cancer
print("Labels = ",cancer.target_names)
```

```python
cancer.data.shape
```

```python
# Print the cancer data features
print(cancer.data[0:5])
```

```python
# Print the cancer labels(0:malignant;1:benign)
print(cancer.target)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,test_size=0.2,random_state=0)
```

```python
# Import svm model
from sklearn import svm

# Create a SVM Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

# Train the model using training set
clf.fit(X_train,y_train)

# Predit the response for test dataset
y_pred= clf.predict(X_test)
```

```python
# Import scikit learn metrics module for accuracy calculation
from sklearn import metrics
score = metrics.accuracy_score(y_test,y_pred)
# Model Accuracy: how often is the classifier correct?
print("Accuracy : ",metrics.accuracy_score(y_test,y_pred)) 
score = metrics.accuracy_score(y_test,y_pred)
```

```python
# Model Precision : what percentage of positive tuples are labeled as such?
print("Precision =  ",metrics.precision_score(y_test,y_pred))

# Model Recall: what percentage of positive tuples are labeled as such?
print("Recall =  ",metrics.recall_score(y_test,y_pred))
```

```python
# Confusion Matrix
cm = metrics.confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(cm,annot=True,fmt='.3f',linewidths=.5,square=True,cmap="Spectral")
plt.ylabel('Actual Label')
plt.xlabel("Predicted Label")
all_sample_title= "Support Vector  Model accuracy (in %):  {0}".format(score*100)
plt.title(all_sample_title,size=15)
```
